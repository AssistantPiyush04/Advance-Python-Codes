```python
#Custom Feature Engineering

from sklearn.preprocessing import FunctionTransformer
import numpy as np

#create a dataset
X=np.array([[1,2],[3,4]])

# define a custom feature Engineering function
def holi(X):
  return np.hstack((X,X**2))

#Create a functionTransformer to apply the custom data
custom_transformer = FunctionTransformer(holi)

#Apply the transformer to the input data
X_transformed= custom_transformer.transform(X)

#view the tranformed data
print(X_transformed)
```

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np

# create a dataset
x=np.array([[1,2],[3,4]])

# define  a custom scalling function
def my_scalling(x):
  return x/np.max(x)

# custom a Function Transformer to apply the custom function
custom_transformer=FunctionTransformer(my_scalling)

x_transformed=custom_transformer.transform(x)

print(x_transformed)
```

```python
## Data Cleaning..

from sklearn.preprocessing import FunctionTransformer
import numpy as np

x=np.array([[1,2],[3,np.nan]])

def my_cleaning(x):
  x[np.isnan(x)]=0
  return x

custom_transformer=FunctionTransformer(my_cleaning)

x_transformed=custom_transformer.transform(x)

print(x_transformed)
```

```python
x=np.array([[df]])

def separate_columns(df):
    int_columns = [col
                   for col in df.columns
                   if df[col].dtype == 'int64']

    float_columns = [col
                     for col in df.columns
                     if df[col].dtype == 'float64']

    object_columns = [col
                      for col in df.columns
                      if df[col].dtype == 'object']

    return int_columns, float_columns, object_columns

int_cols, float_cols, object_cols=separate_columns(df)

print("Integer columns:", int_cols)
print("Float columns:", float_cols)
print("Object columns:", object_cols)



# df.isnull().sum()

```
